Recreate the following plot, and interpret what you see in context of
the data.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
data = {
    "continent": ["Asia", "Europe", "Africa", "North America", "South America", "Oceania"],
    "plastic_waste_per_cap": [0.121, 0.08, 0.144, 0.093, 0.183, 0.112],
    "coastal_pop": [262892387, 4747957, 16556580, 11846863, 16449245, 17235954],
    "total_pop": [1341335152, 10712066, 35468208, 34016593, 40412376, 22268384]
}

df = pd.DataFrame(data)

# Calculate the coastal population proportion
df['coastal_pop_proportion'] = df['coastal_pop'] / df['total_pop']

# Plot
plt.figure(figsize=(10, 8))
sns.scatterplot(x='coastal_pop_proportion', y='plastic_waste_per_cap', hue='continent', data=df)
sns.regplot(x='coastal_pop_proportion', y='plastic_waste_per_cap', data=df, scatter=False, color='black')

plt.title('Plastic waste vs. coastal population proportion\nby continent')
plt.xlabel('Coastal population proportion (Coastal / total population)')
plt.ylabel('Plastic waste per capita')
plt.legend(title='Continent')
plt.show()
```
